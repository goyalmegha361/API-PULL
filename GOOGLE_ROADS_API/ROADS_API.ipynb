# Solution using Wikipedia API

```python
import wikipedia
from bs4 import BeautifulSoup
```

```python
#Search using wikipedia API
search_results=wikipedia.search("Roads in Delhi", results=1)
search_results
```

```python
#Extract html page using beautiful soup
fullpage = wikipedia.page(search_results[0]).html()
print(BeautifulSoup(fullpage))
```

```python
#Extract all the tables using Beautiful soup
roads = pd.DataFrame()
result = BeautifulSoup(fullpage)
tables = result.findAll("table", {"class": "wikitable"})
```

```python
#Extract table contents and append to a dataframe
td=[]
for table in tables:
    #print("###############")
    th = table.find('th', text='Road')
    #print(th)
    td = th.findAllNext('td')
    #print(td)
    for i in range(len(td)):
        b=td[i].text
        print(b)
        roads = roads.append({'Roads' : b},ignore_index=True)
```

```python
#Cleaning the roads table obtained for special characters
sub = roads[0::3]
sub['Roads']=sub['Roads'].str.rstrip("\n")
sub=sub.reset_index(drop=True)
sub
```
